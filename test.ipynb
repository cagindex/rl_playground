```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata

# Example height_data with irregular grid
height_data = {
    (0, 0): 1,
    (0, 2): 3,
    (1, 1): 5,
    (2, 0): 7,
    (2, 2): 9,
    (3, 1): 6,
    (4, 0): 4,
    (4, 2): 2,
    # Add more data points as needed
}

# Extract x, y, z
x = np.array([key[0] for key in height_data.keys()])
y = np.array([key[1] for key in height_data.keys()])
z = np.array(list(height_data.values()))

# Define grid resolution
grid_x, grid_y = np.mgrid[min(x):max(x):100j, min(y):max(y):100j]

# Perform interpolation
grid_z = griddata((x, y), z, (grid_x, grid_y), method='cubic')

# Create the surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(grid_x, grid_y, grid_z, cmap='viridis', edgecolor='none')

# Add color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Height (z)')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (Height)')
ax.set_title('3D Surface Plot with Interpolated Data')

# Adjust the view angle
ax.view_init(elev=30, azim=45)

# Show the plot
plt.show()
```
